```python
#gender
df['gender'] = df.gender.map({1 : 'Male', 0 : 'Female'})

#race
df['race'] = df.race.map({1: 'Black', 2: 'White', 3: 'Hispanic', 4: 'Asian', 6: 'Other'})

# race of the partner
df['race_o'] = df.race_o.map({1: 'Black', 2: 'White', 3: 'Hispanic', 4: 'Asian', 6: 'Other'})

# filed studied
df['field_cd'] = df.field_cd.map({1: 'Law', 2: 'Math', 3: 'Soc. Sc.', 4: 'Med. Sc.', 5: 'Eng.', 6: 'Journ.', 7: 'Hist.', 8: 'Econ', 9: 'Educ.', 10: 'Nat. Sc.', 11: 'Soc. Wr.', 12: 'Und.', 13: 'Pol. Sc.',
14: 'Film', 15: 'Arts', 16:'Lang.', 17: 'Arch.', 18: 'Oth.'})
```

```python
# Proportion du gender
sns.set_theme(style="darkgrid")
ax = sns.countplot(x = df["gender"])
ax.set_ylabel("People quantity")
ax.set_title("Repartition by gender")
print("On constate que la parité est présente dans ce dataset.")
```

```python
plt.subplot(121)
sns.set_theme(style="darkgrid")
ax = sns.countplot(x = df["gender"], palette=['#99ff99','#66b3ff'])
ax.set_ylabel("People quantity")
ax.set_title("Quantity by gender")

plt.subplot(122)
plt.pie(Pie, labels = Lab, autopct='%1.1f%%', colors = color, shadow=True, startangle=60)
plt.title("Male, Female ratio:")
plt.show()
```

```python
# Proportion des race
ax = sns.countplot(x = df["race"], order = ["White", "Asian", "Hispanic", "Other", "Black"])
ax.set_ylabel("People quantity")
ax.set_title("Repartition by races")
print("On constate que la race blanche est sur-représentée dans ce dataset")
```

```python
#répartition des ages :
ax = sns.countplot(x = df["age"])
ax.set_ylabel("How many people")
ax.set_title("Repartition of the people's ages")
print("On constate une majorité de 22/30 ans.")
```

```python
ax = sns.countplot(x = df["field_cd"])
ax.set_xticklabels(ax.get_xticklabels(), 
                    fontsize=15, 
                    rotation=90)
ax.set_title("The main field study", fontsize = 25)
ax.set_ylabel("How many people", fontsize=15)
ax.set_xlabel("Main Field", fontsize=15)
print("On constate que les deux principaux domaines d'études sont l'économie et les sciences naturelles.")
```

```python
#On visualise la réparition des genres par domaines d'études :
df_field_gender = df[['gender', 'field_cd']].groupby(['field_cd', 'gender']).size().unstack()

ax = df_field_gender.plot(kind='bar', 
                    figsize=(12,6), 
                    stacked=True)
ax.set_xticklabels(ax.get_xticklabels(), 
                    fontsize=12, 
                    rotation=45)
ax.set_title("Domaine d'étude par genre", fontsize=20)
ax.set_xlabel("Domaine d'étude",fontsize=15)
ax.set_ylabel("Nombre de pers.",fontsize=15)
plt.show()
print("On constate un déséquilibre du genre dans presque tous les domaines d'étude")
```

```python
#Intérêt pour le sport :
df_field_gender = df[['gender', 'sports']].groupby(['sports', 'gender']).size().unstack()

ax = df_field_gender.plot(kind='bar', 
                    figsize=(15,10), 
                    stacked=True)
ax.set_xticklabels(ax.get_xticklabels(), 
                    fontsize=12, 
                    rotation=0)
ax.set_title("Intéret pour le sport noté de 0 à 10", fontsize=20)
ax.set_xlabel("Note /10",fontsize=15)
ax.set_ylabel("Nombre de pers.",fontsize=15)
plt.show()
print("Pour les notes de 0 à 6 les femmes sont plus représentées et c'est l'inverse de 7 à 10")
```

```python
#On visualise la réparition des renctontres en fonction des genres :
df_field_gender = df[['race', 'race_o']].groupby(['race_o', 'race']).size().unstack()

ax = df_field_gender.plot(kind='bar', 
                    figsize=(15,10), 
                    stacked=True)
ax.set_xticklabels(ax.get_xticklabels(), 
                    fontsize=12, 
                    rotation=0)
ax.set_title("Rencontre en fonction des races", fontsize=20)
ax.set_xlabel("race",fontsize=15)
ax.set_ylabel("Nombre de pers.",fontsize=15)
plt.show()
print("On observe une régularité dans les proportions des rencontres en fonction des races")
print("On peut donc conclure que la race à peu d'impact sur les rencontres")
```
